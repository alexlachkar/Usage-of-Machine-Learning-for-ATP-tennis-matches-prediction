# Create Matches with winner

```python
def matches_stats(data):     
    
    global match_stat
    row=[]
    
    for i in tqdm(range(data.shape[0])):
        
        #info relevant to the tournament: 
        
        tourney_name = data['tourney_name'][i]
        surface = data['surface'][i]
        draw_size = data['draw_size'][i]
        tourney_level = data['tourney_level'][i]
        tourney_date = data['tourney_date'][i]
        Round = data['round'][i]
        minutes = data['minutes'][i]
                
        
#for simplicity the player_1 is defined as the winner
        if data['winner_rank'][i] < data['loser_rank'][i]:
            player_1_rank = data['winner_rank'][i]
            player_1_id = data['winner_id'][i]   
            player_1_name = data['winner_name'][i]
            
            player_2_rank = data['loser_rank'][i]
            player_2_id = data['loser_id'][i] 
            player_2_name = data['loser_name'][i]
            winner = 1
            
        else:
            player_2_rank = data['winner_rank'][i]
            player_2_id = data['winner_id'][i]
            player_2_name = data['winner_name'][i]
            
            player_1_rank = data['loser_rank'][i]
            player_1_id = data['loser_id'][i]  
            player_1_name = data['loser_name'][i]
            winner = 2


        new_row = {'Tourney Name': tourney_name, 'Tourney Date': tourney_date, 'Tourney Level': tourney_level,
                   'Surface': surface, 'Draw size': draw_size, 'Round': Round, 'Minutes': minutes, 
                    'Player1':player_1_name, 'Player1_ID': player_1_id,'Player1_rank': player_1_rank,
                   'Player2':player_2_name,'Player2_ID': player_2_id,'Player2_rank': player_2_rank, 
                      'Winner': winner}
        row.append(new_row)

        match_stat = pd.DataFrame(row, columns = ['Tourney Name','Tourney Date', 'Tourney Level', 'Surface', 'Draw size',
                                                  'Round', 'Minutes',
                                                  'Player1','Player1_ID','Player1_rank',
                                                  'Player2','Player2_ID','Player2_rank', 
                                                 'Winner']) 
        

        
    return (match_stat)
            
            
```

# Include Players' stats to matches

```python
def subtract_days(date, num_days):
    """
    Subtract n days from a specified date
    """
    
    date_temp = date - timedelta(days=num_days)
    return date_temp
```

```python
last_n_weeks = 52 
current_date = pd.to_datetime('06/02/1998')
last_date = subtract_days(current_date, (last_n_weeks * 7))
last_date
```

```python
def statoneplayer(player_id,data):
    
    i = player_id
    
    row = []
    
    wins = 0 
    loses = 0
        
    ace = 0
    svpt = 0
    df = 0 
    firstIn = 0 
    firstWon = 0
    secondWon = 0
    bpSaved = 0 
    bpFaced = 0
        
 #Loop to collect the stats for a player that won a match###
    for j in range(data.shape[0]):
        if (i == data['winner_id'][j]):

            ace = ace + data['w_ace'][j] 
            svpt = svpt + data['w_svpt'][j] 
            df = df + data['w_df'][j] 
            firstIn = firstIn + data['w_1stIn'][j] 
            firstWon = firstWon + data['w_1stWon'][j] 
            secondWon = secondWon + data['w_2ndWon'][j] 
            bpSaved = bpSaved + data['w_bpSaved'][j]
            bpFaced = bpFaced + data['w_bpFaced'][j]

            #ADD: 
            # - AGE avg
            # - RANK avg 

            wins = wins + 1 

#Loop to collect the stats for a player that lost a match#

    for j in range(data.shape[0]):
        if (i == data['loser_id'][j]):

            ace = ace + data['l_ace'][j] 
            svpt = svpt + data['l_svpt'][j] 
            df = df + data['l_df'][j] 
            firstIn = firstIn + data['l_1stIn'][j] 
            firstWon = firstWon + data['l_1stWon'][j] 
            secondWon = secondWon + data['l_2ndWon'][j] 
            bpSaved = bpSaved + data['l_bpSaved'][j]
            bpFaced = bpFaced + data['l_bpFaced'][j]     

            loses = loses + 1   

# Calculate the percentage statistics 

    per_ace = ace/svpt * 100
    per_df = df/svpt * 100
    per_fsM = firstIn/svpt * 100 
    per_fsW = firstWon/firstIn * 100
    per_ssM = (svpt - firstIn)/svpt * 100 
    # per_ssM1 = 100 - per_fsM check is good
    per_ssW = secondWon/(svpt-firstIn) *100
    per_bpSaved = bpSaved/bpFaced * 100        

    total = wins + loses

    if total <2:
        win_percent = 0

    else:
        win_percent = wins / total * 100        


    global player_stat

    new_row = {'Player_ID': i, 'Winning_%': win_percent, 
           'Ace':per_ace,'DoubleFault':per_df,'FirstServeMade':per_fsM,'FirstServeWon':per_fsW,
           'SecondServeMade': per_ssM, 'SecondServeWon':per_ssW,'BreakPointSaved':per_bpSaved}

    row.append(new_row)

    player_stat = pd.DataFrame(row, columns = ['Player_ID','Winning_%','Ace','DoubleFault','FirstServeMade','FirstServeWon',
                                     'SecondServeMade','SecondServeWon','BreakPointSaved']) 
        
    return player_stat
```

```python
def combined_stats_2players(data,match): 
    
    global dataset
    global match_with_stats
    # match_with_stats = match
    allstats2 = pd.DataFrame()
    allstats1 = pd.DataFrame()
    
    for i in tqdm(range(match.shape[0])):      
    
# Filter the n weeks matches        
        last_n_week = 52 #define the number of week you want your stats to be historical 
        match_date = match['Tourney Date'][i]
        last_date = subtract_days(match_date, (last_n_weeks * 7))
        
        dataset = data[(data['tourney_date'] <= match_date) & (data['tourney_date'] >= last_date)]  
        dataset = dataset.reset_index(drop=True)
        
# Starts Calculation 


        j = match['Player1_ID'][i]
        stats1 = statoneplayer(j,dataset)
        allstats1 = allstats1.append(stats1,ignore_index=True) 
        
        k = match['Player2_ID'][i]
        stats2 = statoneplayer(k,dataset)
        allstats2 = allstats2.append(stats2,ignore_index=True)
        
        
    allstats2 = allstats2.rename(columns = {'Player_ID':'Player_ID2',
                                            'Winning_%': 'Winning_%2','Ace':'Ace2','DoubleFault':'DoubleFault2',
                                            'FirstServeMade':'FirstServeMade2','FirstServeWon':'FirstServeWon2',
                                       'SecondServeMade':'SecondServeMade2','SecondServeWon': 'SecondServeWon2','BreakPointSaved':'BreakPointSaved2'})
    
        
    match_with_stats = pd.concat([match, allstats1, allstats2], axis = 1) 
        

    match_with_stats.to_csv(r'C:\Users\alexl\OneDrive\Alex - ESCP\THESIS\Code\match_with_stats.csv',index=False)   
    return match_with_stats
```
